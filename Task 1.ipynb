13. Remove the outliers using IQR and recalculate IQR in outlier removed 'Experience'
column and analyse with boxplot (Use df.copy())

```python
print("J STANES JENSON - URK22CS2061")
df_no_outliers = df.copy()
Q1 = df['Experience'].quantile(0.25)
Q3 = df['Experience'].quantile(0.75)
IQR = Q3 - Q1
lower_limit = Q1 - (1.5 * IQR)
upper_limit = Q3 + (1.5 * IQR)
df_no_outliers = df[(df['Experience'] >= lower_limit) & (df['Experience'] <= upper_limit)]
Q1_no_outliers = df_no_outliers['Experience'].quantile(0.25)
Q3_no_outliers = df_no_outliers['Experience'].quantile(0.75)
IQR_no_outliers = Q3_no_outliers - Q1_no_outliers
print("IQR after removing outliers:", IQR_no_outliers)
plt.boxplot(df_no_outliers['Experience'])
plt.show()
```

14. Remove the outliers using z-score and recalculate z-score in outlier removed
'Experience' column and analyse with boxplot (Use df.copy())

```python
print("J STANES JENSON - URK22CS2061")
df_no_outliers_z = df.copy()
z_scores = stats.zscore(df_no_outliers_z['Experience'])
df_no_outliers_z = df_no_outliers_z[(z_scores < 3) & (z_scores > -3)]
z_scores_no_outliers = stats.zscore(df_no_outliers_z['Experience'])
plt.boxplot(df_no_outliers_z['Experience'])
plt.show()
```
